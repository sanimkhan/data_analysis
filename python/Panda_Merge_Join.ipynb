```python
# Join with index

df1 = pd.read_csv(file)
df2 = pd.read_csv(file2)


df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_L',rsuffix = '_R', how = 'outer')
df4
```

```python
# Same task done using merge

df1 = pd.read_csv(file)
df2 = pd.read_csv(file2)

df1 = df1.set_index('FellowshipID')
df2 = df2.set_index('FellowshipID')

df1.merge(df2, left_index=True, right_index=True, how="outer", suffixes=('_L', '_R'))
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], join = 'inner')
```

```python
pd.concat([df1,df2], join = 'outer')
```

```python
pd.concat([df1,df2], join = 'outer', axis = 1)
```
