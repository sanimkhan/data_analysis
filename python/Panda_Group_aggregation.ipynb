```python
import pandas as pd

file_path = '/Users/sanimkhan/Projects/my-git/data_analysis/python/data/Flavors.csv'
df = pd.read_csv(file_path)


df
```

```python
df2 = df.set_index(['Base Flavor', 'Flavor'])
df2 = df2.sort_index(inplace=False)
df2[['Liked', 'Flavor Rating']]
```

```python
# group by
gp = df.groupby('Base Flavor')
numeric_columns = df.select_dtypes(include=['number']).columns
gp[numeric_columns].mean()
```

```python
df.groupby('Base Flavor')[numeric_columns].mean()
```

```python
df.groupby('Base Flavor').count()
```

```python
gp[numeric_columns].sum()
```

```python
# Group by Multiple Columns
gp1 = df.groupby(['Base Flavor', 'Liked'])
gp1[numeric_columns].mean()
```

```python
#aggregation
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean','max','count','sum'], 'Texture Rating':['mean','max','count','sum']})
```

```python
# Transforming Data
df['Normalized Rating'] = df.groupby('Base Flavor')['Total Rating'].transform(lambda x: x / x.max())
df[['Flavor', 'Base Flavor', 'Total Rating', 'Normalized Rating']]
```
